```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
```

```python
df=pd.read_csv("fi.csv")
df
```

```python
X=df.iloc[:,1:]
Y=df.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split as t
xtest,xtrain,ytest,ytrain=t(X,Y,test_size=0.8)
```

```python
from sklearn.preprocessing import StandardScaler as s
sc=s()
X=sc.fit_transform(X)
xtrain=sc.transform(xtrain)
xtest=sc.transform(xtest)
```

```python
from sklearn.decomposition import PCA as p
pc=p(n_components=2)
yt=pc.fit_transform(X)
```
